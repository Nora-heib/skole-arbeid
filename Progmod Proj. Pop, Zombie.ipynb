```python
from pylab import * 

#Initialverdier
t0 = 0 #Starttid i månder
M0 = 500.0 #Antall mennesker ved starten
Z0 = 200 #Startantall zombier
inn = uniform(0,0.04) # innvandring
f = uniform(0.01, 0.028) + inn #Reproduksjonskode, Jeg antar at en trejedel har barn en gang i året, siden de vil ha flere folk her er også innvandring i prosent, for desto flere som bor der, desto flere vil komme inn
b = 1000 #Bæreevne til menneskene
zb = 700 #Bæreevne til zombiene (tenker de dreper hverandre når det er for mange av de)
zm = uniform(0.00002,0.01) #Møteraten for en zombie å møte et menneske
mz = uniform(0.001, 0.1) #Møteraten for et menneske å møte en zombie
d = uniform(0,0.2) # Hvor mange % av zombiene som blir drept av mennesker
a = uniform(0.05, 0.1) #hvor mange mennesker som dør av alder, sykdom eller skade
zi = -(uniform(0.01, 0.5)) #Innvandringen av zombier fra andre områder
zombarn = 0.01
# Tidssteg
N = 100000 # antall intervall
tid = 100 # Antall måneder man føler med
dt = tid/(N-1) #tidssteg
AH = randint(1,10) #Prosent hjelpen fra andre byer tar å dreper zombiene hvis populasjonen går for lavt
AH_plus = 30 #Hvor mange mennesker som flytter inn i byen hvis det blir for få mennesker

#Matriser
t = zeros(N) #Tiden
H = zeros(N) #Formelen for menneskeheten i den vanlige grafen
Hder = zeros(N) # Den deriverte til formelen til menneskeheten i den vanlige grafen
Z = zeros(N) # Formel for zombiene i den vanlige grafen
Zder = zeros(N) #formel for den deriverte i den vanlige grafen
bH = zeros(N) #Formelen for menneskeheten i den beste sit grafen
bHder = zeros(N)#formel for den deriverte av mennesker i den beste sit grafen
bZ = zeros(N)#Formelen for zombier i den beste sit grafen
bZder = zeros(N)#formel for den deriverte av zombier i den beste sit grafen
vH  = zeros(N)#Formelen for menneskeheten i den værste sit grafen
vHder = zeros(N)#formel for den deriverte av mennesker i den vørste sit grafen
vZ = zeros(N)#Formelen for zombier i den værste sit grafen
vZder = zeros(N)#formel for den deriverte av zombier i den værste sit. grafen


#Initierer matrisene
t[0] = t0
H[0] = M0
Z[0] = Z0
bH[0] = M0
bZ[0] = Z0
vH[0] = M0
vZ[0] = Z0


# Eulers metode for mest vanlig situasjon
for i in range(N-1): 
    Hder[i] = f * H[i] * (1 -(H[i]/b)) - (H[i] * mz * 0.2) - (H[i]*a) + (inn * H[i]) # fødselsraten ganger populasjonen ganger bærekraften minus hvor mange som dør pga. zombier når de går ut etter nye ressuser
    Zder[i] =(1 - (Z[i]/zb)) * Z[i] + ((H[i] * Z[i] * mz * 0.2) * 0.6) - (H[i]* d) #Bæreevnen til zombiene plus prosenten av mennesker som blir zombier plus hvor mange mennesker som dreper zombier
    H[i + 1] = H[i] + Hder[i] * dt #Formelen for mennesker
    Z[i + 1] =  Z[i] + Zder[i] * dt #formelen for zombier
    t[i + 1] = t[i] + dt #Endringen i tid

    if H[i] < 100: #Hvis det blir for få zombier så skal det komme hjelp fra utsiden. Dette er det 
        Z[i + 1] = Z[i] - (0.001 * AH) * Z[i] #hvor mange prosent zombier blir drept
        H[i + 1] = AH_plus + H[i] #Det kommer flere folk inn

    if Z[i] <= 5: #Hvis det blir for få zombier så flytter det nye inn
        Z[i+1] = 5 + Z[i] * zi  


# Eulers metode for beste situasjonen. Alt er likt untatt det er en kur for zombiene. Menneskene gjør så mye so mulig, og kurer 90% av zombiene de møter
for i in range(N-1): 
    bHder[i] = f * bH[i] * (1 -(bH[i]/b)) + (bH[i] * mz) - (0.06 * zm * bH[i]) - (bH[i]*a)  # fødselsraten ganger populasjonen ganger bærekraften minus hvor mange som dør pga. alder. De kurer alle zombiene de møter. noen dør.  ikke så mange 
    bZder[i] = (1 - (bZ[i]/zb)) * bZ[i]  - (bH[i] * bZ[i] * zm * 0.9) #Bæreevnen til zombiene. Ingen mennesker blir zombioer siden de kan bli kurert minus hvor mange zombier som blir mennesker igjen.
    bH[i + 1] = bH[i] + bHder[i] * dt #Formelen for mennesker
    bZ[i + 1] =  bZ[i] + bZder[i] * dt #formelen for zombier
    t[i + 1] = t[i] + dt #Endringen i tid

    if bH[i] < 100: #Hvis det blir for få zombier så skal det komme hjelp fra utsiden. Dette er det 
        bZ[i + 1] = bZ[i] - (0.001 * AH) * bZ[i] #hvor mange prosent zombier blir drept
        bH[i + 1] = AH_plus + bH[i] #Det kommer flere folk inn
        
    if bZ[i] <= 5: #Hvis det blir for få zombier så flytter det nye inn
        bZ[i+1] = 5 + bZ[i] * zi  

       
    # Eulers metode for værste situasjon. Her er zombiene faligre og klarer å gjøre om alle menneskene som de fanger til zombier, som er dålig. Menneskene er like flinke til å¨drepe de, siden hvis det er en ting folk er flinke til, så er det å drepe ting. Mennesker får de ikke noe hjelp fra utsidne heller
# Det værste med denne er at de kan ha barn.... Zombiene kan ha barn...
for i in range(N-1): 
    vHder[i] = f * vH[i] * (1 -(vH[i]/b)) - (vH[i] * mz * 0.1) - (vH[i]*a) + (inn * vH[i])  # fødselsraten ganger populasjonen ganger bærekraften minus hvor mange som dør pga. zombier når de går ut etter nye ressuser
    vZder[i] =(1 - (vZ[i]/zb)) *vZ[i] + vZ[i]*zombarn + (vH[i] * mz * 0.1) - (vH[i]* d) #Bæreevnen til zombiene plus prosenten av mennesker som blir zombier plus hvor mange mennesker som dreper zombier
    vH[i + 1] = vH[i] + vHder[i] * dt #Formelen for mennesker
    vZ[i + 1] =  vZ[i] + vZder[i] * dt #formelen for zombier
    t[i + 1] = t[i] + dt #Endringen i tid

    if vZ[i] <= 5: #Hvis det blir for få zombier så flytter det nye inn
        vZ[i+1] = 5 + vZ[i] * zi



#Grafen med vanlig situasjon        
plot(t,H)
plot(t, Z, color = 'forestgreen')
title('Zombie apocalypse')
xlabel('tid')
ylabel('Antall')
legend(['Pop. Mennesker', 'Pop. Zombier'])
show()

#Grafen som viser den beste situasjonen
plot(t,bH)
plot(t, bZ, color = 'forestgreen')
title('Beste Zombie apocalypse')
xlabel('tid')
ylabel('Antall')
legend(['Pop. Mennesker', 'Pop. Zombier'])
show()

#Grafen med den værste situasjonen
plot(t,vH)
plot(t, vZ, color = 'forestgreen')
title('Værste Zombie apocalypse')
xlabel('tid')
ylabel('Antall')
legend(['Pop. Mennesker', 'Pop. Zombier'])
show()
```
